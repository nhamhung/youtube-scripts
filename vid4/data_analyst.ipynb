### Phân tích số lượng tỷ phú theo lĩnh vực (Industry)

```python
industry_analysis = data.groupby('industries').agg({
    'personName': 'count',
    'finalWorth': 'sum'
}).rename(columns={'personName': 'Number of Billionaires', 'finalWorth': 'Total Net Worth (in millions)'})

# Sorting the industries based on the number of billionaires
industry_analysis = industry_analysis.sort_values(by='Number of Billionaires', ascending=False)

industry_analysis
```

```python
# Setting up the figure for subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 15))

# Plotting Number of Billionaires
industry_analysis['Number of Billionaires'].plot(kind='bar', ax=axes[0], color='blue', alpha=0.6)
axes[0].set_title('Number of Billionaires by Industry')
axes[0].set_ylabel('Number of Billionaires')
axes[0].set_xlabel('Industry')

# Plotting Total Net Worth (in millions)
industry_analysis['Total Net Worth (in millions)'].plot(kind='bar', ax=axes[1], color='green', alpha=0.6)
axes[1].set_title('Total Net Worth by Industry (in millions)')
axes[1].set_ylabel('Total Net Worth (in millions)')
axes[1].set_xlabel('Industry')

plt.tight_layout()
plt.show()
```

### Phân tích số lượng tỷ phú theo đất nước (Country)

```python
# Grouping data by country to analyze which countries have the most billionaires and their combined net worth
country_analysis = data.groupby('country').agg({
    'personName': 'count',
    'finalWorth': 'sum'
}).rename(columns={'personName': 'Number of Billionaires', 'finalWorth': 'Total Net Worth (in millions)'})

# Sorting the countries based on the number of billionaires
country_analysis = country_analysis.sort_values(by='Number of Billionaires', ascending=False)

country_analysis.head(10)  # Displaying top 10 countries for brevity
```

```python
# Setting up the figure for subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 15))

# Plotting Number of Billionaires for the top 10 countries
country_analysis['Number of Billionaires'].head(10).plot(kind='bar', ax=axes[0], color='blue', alpha=0.6)
axes[0].set_title('Top 10 Countries by Number of Billionaires')
axes[0].set_ylabel('Number of Billionaires')
axes[0].set_xlabel('Country')

# Plotting Total Net Worth (in millions) for the top 10 countries
country_analysis['Total Net Worth (in millions)'].head(10).plot(kind='bar', ax=axes[1], color='green', alpha=0.6)
axes[1].set_title('Total Net Worth by Country (Top 10) in millions')
axes[1].set_ylabel('Total Net Worth (in millions)')
axes[1].set_xlabel('Country')

plt.tight_layout()
plt.show()
```

Các đất nước với nhiều tỉ phú nhất == Các đất nước với tổng tài sản của các tỷ phú cao nhất:
1. Mỹ
2. Trung
3. Ấn

### Phân tích tương quan (Correlation Analysis)

```python
# Calculating the correlation matrix for the dataset
correlation_matrix = data.corr()

correlation_matrix
```

```python
import seaborn as sns

# Plotting a heatmap of the correlation matrix to visualize the most meaningful correlations
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, annot=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Phân tích phân phối theo Độ tuổi, Tài sản ròng (Age / Net Worth Distribution)

```python
# Setting up the figure for subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 15))

# Plotting the distribution for 'finalWorth' (Net Worth)
sns.histplot(data['finalWorth'], bins=50, ax=axes[0], color='skyblue', kde=True)
axes[0].set_title('Net Worth Distribution (in millions)')
axes[0].set_xlabel('Net Worth (in millions)')
axes[0].set_ylabel('Number of Billionaires')

# Plotting the distribution for 'age'
sns.histplot(data['age'].dropna(), bins=50, ax=axes[1], color='salmon', kde=True)
axes[1].set_title('Age Distribution')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Number of Billionaires')

plt.tight_layout()
plt.show()
```

Tài sản ròng:
- Phần lớn các tỷ phú có giá trị tài sản ròng ở mức thấp hơn (khoảng 1 tỷ USD đến 10 tỷ USD).
- Càng có ít tỷ phú hơn khi giá trị tài sản ròng tăng lên, điều này cho thấy giá trị tài sản ròng cực cao càng hiếm hơn.
- Sự phân bổ bị lệch phải, có nghĩa là có một số tỷ phú có tài sản ròng cực kỳ cao, điều này kéo giá trị trung bình sang phải.

Tuổi:
- Sự phân bổ độ tuổi có vẻ bình thường, với phần lớn các tỷ phú ở độ tuổi từ 50 đến 70.
- Có ít tỷ phú trẻ (dưới 40) và tỷ phú lớn tuổi (trên 80).
- Đỉnh điểm cho thấy độ tuổi phổ biến nhất của các tỷ phú là khoảng đầu những năm 60.
